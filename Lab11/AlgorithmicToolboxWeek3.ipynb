## **1. Money Exchange**
In this problem, you will design and implement an elementary greedy algorithm used by cashiers all over the world millions of times per day.
### **Problem Description**
**Task.**  The goal in this problem is to find the minimum number of coins needed to change the input value (an integer) into coins with denominations 1, 5, and 10.

**Input Format.** The input consists of a single integer *m*.

**Constraints.** 0 ‚â§ *m* ‚â§ $10^3$.

**Output Format.** Output the minimum number of coins with denominations 1, 5, 10 that changes ùëö.
### **Sample 1**
Input:

 2

2
### **Sample 2**
Input:

28

Output: 

6


### **Code**

```python
%%writefile change.cpp
#include <iostream>
#include <vector>
int get_change(int m){
  int value=m;
  int n=0;

  while (m!=0){
      if ((m-10)>=0){
          n++;
          m=m-10;

      }
      else if((m-5)>=0){
           n++;
           m=m-5;
      }
      else{
           n++;
           m--;
      }
  }

  return n;
}

int main() {
  int m;
  std::cin >> m;
  std::cout << get_change(m) << '\n';
}
```

## **2. Maximum value of the loot**
A thief finds much more loot than his bag can fit. Help him to find the most valuable combination of items assuming that any fraction of a loot item can be put into his bag
### **Problem Description**
**Task.**  The first line of the input contains the number ùëõ of items and the capacity ùëä of a knapsack. The next ùëõ lines define the values and weights of the items. The ùëñ-th line contains integers ùë£ùëñ and ùë§ùëñ‚Äîthe value and the weight of ùëñ-th item, respectively.

**Input Format.** The first line of the input contains the number ùëõ of items and the capacity ùëä of a knapsack.
The next ùëõ lines define the values and weights of the items. The ùëñ-th line contains integers ùë£ùëñ and ùë§ùëñ‚Äîthe
value and the weight of ùëñ-th item, respectively.


**Constraints.** 1 ‚â§ ùëõ ‚â§ $10^3$.
, 0 ‚â§ ùëä ‚â§ 2 ¬∑ $10^6$.
; 0 ‚â§ ùë£ùëñ ‚â§ 2 ¬∑ $10^6$.
, 0 < ùë§ùëñ ‚â§ 2 ¬∑ $10^6$.
for all 1 ‚â§ ùëñ ‚â§ ùëõ. All the
numbers are integers

**Output Format.** Output the maximal value of fractions of items that fit into the knapsack.
### **Sample 1**
Input:

3 50
60 20
100 50
120 30

Output: 

180.0000
### **Sample 2**
Input:

1 10
500 30

Output: 

166.6667


### **Code**

```python
%%writefile loot.cpp
#include <iostream>
#include <vector>

using std::vector;

double get_optimal_value(int capacity, vector<int> weights, vector<int> values) {
  double value = 0.0;
  int full= capacity;
  int l= values.size();
  vector<double> value_weight(l);
  int maxValue=0;

  for(int i=0;i<l;i++){
      value_weight[i]=((double)values[i])/((double)weights[i]);

  }
  

while(capacity!=0){
  maxValue=0;
  for(int j=0;j<l;j++){
      if(value_weight[j]>=value_weight[maxValue]){
          maxValue=j;
       }
    }
  if(weights[maxValue]>=capacity){
        value= value + value_weight[maxValue]*capacity;
        capacity=0;
    }
    else {
        value=value+ value_weight[maxValue]*weights[maxValue];
        capacity=capacity-weights[maxValue];
        value_weight[maxValue]=0;
    }
  }


  return value;
}

int main() {
  int n;
  int capacity;
  std::cin >> n >> capacity;
  vector<int> values(n);
  vector<int> weights(n);
  for (int i = 0; i < n; i++) {
    std::cin >> values[i] >> weights[i];
  }

  double optimal_value = get_optimal_value(capacity, weights, values);

  std::cout.precision(10);
  std::cout << optimal_value << std::endl;
  return 0;
}
```

## **3. Car Fueling**
You are going to travel to another city that is located ùëë miles away from your home city. Your car can travel at most ùëö miles on a full tank and you start with a full tank. Along your way, there are gas stations at distances stop1,stop2, . . . ,stopùëõ from your home city. What is the minimum number of refills needed?
### **Problem Description**
**Task.**  The first line contains an integer ùëë. The second line contains an integer ùëö. The third line specifies an integer ùëõ. Finally, the last line contains integers stop1 ,stop2 , . . . ,stopùëõ

**Input Format.** The first line of the input contains the number ùëõ of items and the capacity ùëä of a knapsack.
The next ùëõ lines define the values and weights of the items. The ùëñ-th line contains integers ùë£ùëñ and ùë§ùëñ‚Äîthe
value and the weight of ùëñ-th item, respectively.


**Constraints.**  1 ‚â§ ùëë ‚â§ 105. 1 ‚â§ ùëö ‚â§ 400. 1 ‚â§ ùëõ ‚â§ 300. 0 < stop1 < stop2 < ¬∑ ¬∑ ¬∑ < stopùëõ < ùëë.

**Output Format.** Assuming that the distance between the cities is ùëë miles, a car can travel at most ùëö miles on a full tank, and there are gas stations at distances stop1,stop2, . . . ,stopùëõ along the way, output the
minimum number of refills needed. Assume that the car starts with a full tank. If it is not possible to reach the destination, output ‚àí1
### **Sample 1**
Input:

950
400
4
200 375 550 750

Output: 

2
### **Sample 2**
Input:

10
3
4
1 2 5 9

-1


### **Code**

```python
%%writefile car_fueling.cpp
#include <iostream>
#include <vector>
using std::cin;
using std::cout;
using std::vector;
using std::max;

int compute_min_refills(int dist, int tank, vector<int> & stops) {
    int stopsNeed=0;
    int tempStop=0;
    int lastStop=0;
    int l= stops.size();

    if(((dist-stops[l-1])>tank)||(stops[0]>tank)){ 
        return -1;
    }
 

    while (tempStop<=l){
        lastStop=tempStop;
        while((tempStop<=l)&&((stops[tempStop+1]-stops[lastStop])<=tank)){
            tempStop++;
        }

       if (lastStop==tempStop){
            return -1;
        }

     
        stopsNeed++;
          
        
    }

    
    return stopsNeed;
    
   
}


int main() {
    int d = 0;
    cin >> d;
    int m = 0;
    cin >> m;
    int n = 0;
    cin >> n;

    vector<int> stops(n);
    for (size_t i = 0; i < n; ++i)
        cin >> stops.at(i);

    cout << compute_min_refills(d, m, stops) << "\n";

    return 0;
}
```

## **6. Maximum Number of Prizes**
You are organizing a funny competition for children. As a prize fund you have ùëõ
candies. You would like to use these candies for top ùëò places in a competition
with a natural restriction that a higher place gets a larger number of candies.
To make as many children happy as possible, you are going to find the largest
value of ùëò for which it is possible.
### **Problem Description**
**Task.**  The goal of this problem is to represent a given positive integer ùëõ as a sum of as many pairwise distinct positive integers as possible. That is, to find the maximum ùëò such that ùëõ can be written as ùëé1 + ùëé2 + ¬∑ ¬∑ ¬∑ + ùëéùëò where ùëé1, . . . , ùëéùëò are positive integers and ùëéùëñ Ã∏= ùëéùëó for all 1 ‚â§ ùëñ < ùëó ‚â§ ùëò.

**Input Format.** The input consists of a single integer ùëõ.

**Constraints.** 1 ‚â§ ùëõ ‚â§ $10^9$



**Output Format.** In the first line, output the maximum number ùëò such that ùëõ can be represented as a sum of ùëò pairwise distinct positive integers. In the second line, output ùëò pairwise distinct positive integers that sum up to ùëõ (if there are many such representations, output any of them).
### **Sample 1**
Input:

6

Output: 

3

1 2 3

### **Sample 2**
Input:

8

Output: 

3

1 2 5
### **Sample 3**
Input:

2

Output: 


1

2

### **Code**

```python
%%writefile df.cpp
#include <iostream>
#include <vector>

using std::vector;

vector<int> optimal_summands(int n) {
  vector<int> summands;
  
  
  for (int k = n, l = 1; k > 0; l++) {
        if (k <= 2 * l) {
            summands.push_back(k);
            k -= k;
        } else {
            summands.push_back(l);
            k -= l;
        }
    }
  return summands;
}

int main() {
  int n;
  std::cin >> n;
  vector<int> summands = optimal_summands(n);
  std::cout << summands.size() << '\n';
  for (size_t i = 0; i < summands.size(); ++i) {
    std::cout << summands[i] << ' ';
  }
}
```
